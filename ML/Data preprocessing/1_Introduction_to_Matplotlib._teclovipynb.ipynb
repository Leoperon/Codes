# Basic Plotting 

Let's start with the basics.

## Basic Plotting, Axes Labels and Titles

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(5, 100, 100)
y = np.linspace(10, 1000, 100)

plt.plot(x, y)
````

````python
plt.show()
````

````python
plt.plot([1, 4, 6, 8], [3, 8, 3, 5])
plt.show()
````


And we can specify the limits of x and y labels as a range using ```xlim([xmin, xmax])``` and ```ylim([ymin, ymax])```.


````python
plt.plot(x, y)
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")
plt.xlim([20, 80])
plt.ylim([200, 800])
plt.show()
````

````python
x = np.linspace(0, 10, 20)
y = x*2
plt.plot(x, y, 'b+')
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")
plt.show()
````

````python
x = np.linspace(0, 5, 10)
y = np.linspace(3, 6, 10)
plt.plot(x, y, 'r-', x, y**2, 'b+', x, y**3, 'g^')
plt.show()
````

## Figures and Subplots


````python
x = np.linspace(1, 10, 100)
y = np.log(x)
plt.figure(1)
plt.subplot(121)     # plt.subplot(1, 2, 1)
plt.title("y = log(x)")
plt.plot(x, y)
plt.subplot(122)
plt.title("y = log(x)**2")
plt.plot(x, y**2)
plt.show()
````

````python
x = np.linspace(1, 10, 100)
plt.figure(1)
plt.subplot(2, 2, 1)
plt.title("Linear")
plt.plot(x, x)
plt.subplot(2, 2, 2)
plt.title("Cubic")
plt.plot(x, x**3)
plt.figure(2)
plt.subplot(2, 2, 1)
plt.title("Log")
plt.plot(x, np.log(x))
plt.subplot(2, 2, 2)
plt.title("Exponential")
plt.plot(x, x**2)
plt.show()
````

## Types of Commonly Used Plots

- Boxplots
- Histograms
- Scatter plots
- Bar plots

````python
df = pd.read_csv("./global_sales_data/market_fact.csv")
df.head()
````

### Boxplot

````python
plt.boxplot(df['Order_Quantity'])
plt.show()
````

````python
plt.boxplot(df['Sales'])
plt.show()
````

````python
df['Sales'].describe()
````

````python
plt.subplot(1, 2, 1)
plt.boxplot(df['Sales'])
plt.subplot(1, 2, 2)
plt.boxplot(df['Sales'])
plt.yscale('log')
plt.show()
````

### Histogram



````python
plt.hist(df['Sales'])
plt.show()
````

````python
plt.hist(df['Sales'])
plt.yscale('log')
plt.show()
````

### Scatter Plot



````python
plt.scatter(df['Sales'], df['Profit'])
plt.show()
````
